```python
from sklearn.datasets import load_breast_cancer
import numpy as np
dataset = load_breast_cancer()
x=dataset.data
y=dataset.target
print(x.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
trn_x, tst_x, trn_y, tst_y = train_test_split(x, y, test_size=0.4, random_state=1234)
print(trn_x.shape, tst_x.shape)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression().fit(trn_x, trn_y)
print("로지스틱 회귀계수", clf.coef_)
```

```python
from sklearn.metrics import accuracy_score
fitt = clf.predict(trn_x)
print(accuracy_score(y_true=trn_y, y_pred=fitt))
pred = clf.predict(tst_x)
print(accuracy_score(y_true=tst_y, y_pred=pred))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_true=tst_y, y_pred=pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(tst_y, pred))
```

```python
path = '/content/drive/MyDrive/AP.csv'
import pandas as pd
df = pd.read_csv(path)
```

```python
names_x = ['G.S',  'T.S',  'U.R',  'SOP',  'LOR',  'CGPA',  'Research','C.o.A']
x = df[names_x]
print(x.shape)
```

```python
import numpy as np
```

```python
from sklearn.decomposition import PCA
```

```python
pca4 = PCA(n_components=8)
pc4 = pca4.fit_transform(x)
stu_id = ['stu'+str(v+1) for v in range(len(x))]
pc_id = ['stu'+str(v+1) for v in range(8)]
pd.DataFrame(pc4, columns=pc_id, index=stu_id)
```

```python
var_pc = np.var(pc4, axis=0)
pd.DataFrame(var_pc, index=pc_id, columns=['Variance'])
```

```python
print("주성분 분석 결과, 총 분산의 합 =", np.sum(var_pc))
var_x = np.var(x, axis=0)
print("원래 데이터 분산의 합 =", np.sum(var_x))
```

```python
from sklearn.preprocessing import StandardScaler
scaled_x = StandardScaler().fit_transform(x)
var_scaled_x = np.var(scaled_x, axis=0)
print("표준화된 데이터 분산의 합 =", np.sum(var_scaled_x))
```

```python
pd.DataFrame(scaled_pc4,columns=pc_id,index=stu_id)
```

```python
path = '/content/drive/MyDrive/AP.csv'
import pandas as pd
df = pd.read_csv(path)

print(df.shape)
print(df.head(5))
```

```python
names_x = ['G.S',  'T.S',  'U.R',  'SOP',  'LOR',  'CGPA',  'Research']
names_y = ['C.o.A']

x = df[names_x]
y = df[names_y]

print(x.shape)
```

```python
df['NY']= df['C.o.A'].apply(lambda x:1 if x>=0.6 else 0)
```

```python
df.head(5)
```

```python
import numpy as np
c = df['NY'].to_list()
freq, uniq = np.unique(c, return_counts=True)

pd.DataFrame({'Frequency': freq,
              'Class':uniq})
```

```python
from sklearn.model_selection import train_test_split
```

```python
trn_x, tst_x, trn_c, tst_c = train_test_split(x,
                                              c,                                              
                                              test_size=0.3,
                                              random_state=1234)
```

```python
from sklearn.linear_model import LogisticRegression
clf_LR = LogisticRegression().fit(trn_x, trn_c)
cls_LR = clf_LR.predict(tst_x)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_true=tst_c, y_pred=cls_LR))
print(classification_report(y_true=tst_c, y_pred=cls_LR))
```

```python
from sklearn.metrics import accuracy_score
acc_LR = accuracy_score(y_true=tst_c, y_pred=cls_LR)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf_DT = DecisionTreeClassifier().fit(trn_x, trn_c)
cls_DT = clf_DT.predict(tst_x)
acc_DT = accuracy_score(y_true=tst_c, y_pred=cls_DT)
```

```python
print(confusion_matrix(y_true=tst_c, y_pred=cls_DT))
print(classification_report(y_true=tst_c, y_pred=cls_DT))
```

```python
print("로지스틱회귀의 정확도: ", acc_LR)
print("의사결정나무의 정확도: ", acc_DT)
```

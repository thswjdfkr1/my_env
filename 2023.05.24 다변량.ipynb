```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x, y = make_blobs(n_samples=1000, n_features=2, centers=3, cluster_std=4, random_state=100)
```

```python
from sklearn.model_selection import train_test_split
trn_x, tst_x, trn_y, tst_y = train_test_split(x, y, test_size=0.4, random_state=1234)
print(trn_x.shape, tst_x.shape)
```

```python
plt.figure(figsize=(5,5))
plt.plot(x[:,0][y==0],x[:,1][y==0], "ro", markersize=2)
plt.plot(x[:,0][y==1],x[:,1][y==1], "go", markersize=2)
plt.plot(x[:,0][y==2],x[:,1][y==2], "bo", markersize=2)
plt.xlabel("x1", fontsize=10)
plt.ylabel("x2", fontsize=10)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
RandomForestClassifier(n_estimators=100, max_depth=3).fit(trn_x, trn_y)
```

```python
from dslearn import multi_stat
```

```python
rf_t1 = RandomForestClassifier(n_estimators=1, max_depth=3).fit(trn_x, trn_y)
multi_stat.viz_boundary(rf_t1, trn_x, trn_y)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth=3).fit(trn_x, trn_y)
multi_stat.viz_boundary(dt, trn_x, trn_y)
```

```python
rf_t50 = RandomForestClassifier(n_estimators=50, max_depth=3).fit(trn_x, trn_y)
multi_stat.viz_boundary(rf_t50, trn_x, trn_y)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
father = np.array([152,168,170,180,183])
son = np.array([163,166,175,178,174])
```

```python
X= father.reshape(-1,1)
y= son.reshape(-1,1)
```

```python
model=LinearRegression().fit(X,y)
print(model.coef_, model.intercept_)
```

```python
from dslearn import multi_stat
```

```python
results= multi_stat.lm_stat(model=model, X=X, y=y)
```

```python
import pandas as pd
```

```python
pd.DataFrame(results)
```

```python
print("결정계수 :", multi_stat.lm_r2(model, X, y, adjust=False))
print("수정된 결정계수 :", multi_stat.lm_r2(model, X, y, adjust=True))
```

```python
pred = model.predict(X)
error = [yy[0] -yy[1] for yy in zip(y,pred)]
np.round(np.concatenate((y, pred, error), axis=1))
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
MAD = mean_absolute_error(y_true=y, y_pred=pred)
MSE = mean_squared_error(y,pred)
RMSE = np.sqrt(MSE)
```

```python
print(MAD, MSE, RMSE)
```
